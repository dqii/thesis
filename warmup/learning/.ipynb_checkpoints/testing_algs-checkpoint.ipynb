```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from algs import logistic, hinge
from learner import evaluate
from display_results import disp_results

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
# DIABETES DATASET, BINARY (768, 9)
filename = '../datasets/pima-indians-diabetes.csv'
dataset = np.genfromtxt(filename, delimiter=',')
np.array(dataset).astype(float)
split = np.split(dataset, [-1], axis=1)
diabetes_x = np.insert(split[0], 0, 1, axis=1)
diabetes_x = MinMaxScaler().fit_transform(diabetes_x)
diabetes_y = np.ndarray.flatten(split[1]).astype(int)
```

```python
# LETTER RECOGNITION, MULTICLASS (20000, 17)
filename = '../datasets/letter-recognition.csv'
dataset = np.genfromtxt(filename, delimiter=',', dtype='<U3')
split = np.split(dataset, [1], axis=1)
letter_x = np.insert(split[1], 0, 1, axis=1).astype('float')
letter_x = MinMaxScaler().fit_transform(letter_x)
convert_to_num = np.vectorize(ord)
letter_y = convert_to_num(np.ndarray.flatten(split[0])) - ord('A')
if min(letter_y) != 0:
    letter_y -= min(letter_y)
```

```python
# WHITE WINE, MULTICLASS (4898, 12)
filename = '../datasets/winequality-white.csv'
dataset = np.genfromtxt(filename, delimiter=';')
np.array(dataset).astype(float)
split = np.split(dataset, [-1], axis=1)
whitewine_x = np.insert(split[0], 0, 1, axis=1)
whitewine_x = MinMaxScaler().fit_transform(whitewine_x)
whitewine_y = np.ndarray.flatten(split[1]).astype(int)
whitewine_y -= min(whitewine_y)
```

```python
# RED WINE, MULTICLASS (1599, 12)
filename = '../datasets/winequality-red.csv'
dataset = np.genfromtxt(filename, delimiter=';')
np.array(dataset).astype(float)
split = np.split(dataset, [-1], axis=1)
redwine_x = np.insert(split[0], 0, 1, axis=1)
redwine_x = MinMaxScaler().fit_transform(redwine_x)
redwine_y = np.ndarray.flatten(split[1]).astype(int)
redwine_y -= min(redwine_y)
```

```python
# MNIST, MULTICLASS (60000, 785)
filename = '../datasets/mnist.csv'
dataset = np.genfromtxt(filename, delimiter=',')
np.array(dataset).astype(int)
split = np.split(dataset, [1], axis=1)
mnist_x = np.insert(split[1], 0, 1, axis=1)
mnist_x = MinMaxScaler().fit_transform(mnist_x)
mnist_y = np.ndarray.flatten(split[0]).astype(int)
```

```python
# evaluation parameters
n_folds = 5
lr = 0.05
n_epoch = 100
```

```python
print("Diabetes Dataset")
x = diabetes_x
y = diabetes_y
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, logistic)
display_results.disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Logistic", LogisticRegression)
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, hinge)
display_results.disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Hinge", SVC)
```

```python
print("Letter Dataset")
x = letter_x
y = letter_y
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, logistic)
disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Logistic", LogisticRegression)
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, hinge)
disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Hinge", SVC)
```

```python
print("White Wine Dataset")
x = whitewine_x
y = whitewine_y
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, logistic)
disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Logistic", LogisticRegression)
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, hinge)
disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Hinge", SVC)
```

```python
print("Red Wine Dataset")
x = redwine_x
y = redwine_y
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, logistic)
disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Logistic", LogisticRegression)
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, hinge)
disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Hinge", SVC)
```

```python
print("MNIST Dataset")
x = mnist_x
y = mnist_y
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, logistic)
disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Logistic", LogisticRegression)
training_accuracy, testing_accuracy, loss, accuracy_while_training = evaluate(x, y, n_folds, lr, n_epoch, hinge)
disp_results(x,y,training_accuracy, testing_accuracy, loss, accuracy_while_training, "Hinge", SVC)
```
