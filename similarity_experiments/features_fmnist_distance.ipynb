```python
%load_ext autoreload
%autoreload 2
import numpy as np
import tensorflow as tf
from features.settings import Triplet_Distance_Settings
from features.models import GCNNMedium0, GCNNMedium1
from learn import get_feature_model, train_features, test_features
from datasets import load_fashion_mnist
```

```python
data, samplers, pair_samplers, triplet_samplers = load_fashion_mnist(augment=True)
```

```python
d = data; s=triplet_samplers
```

# 10

```python
tf.reset_default_graph()
f, f_params = get_feature_model(feature_model=GCNNMedium0, settings=Triplet_Distance_Settings,
                                num_features=10, data_format=[28, 28, 1],
                                f_lr=0.03, f_reg=0.001, f_scope="d_gm0_fmnist_10")
fs = tf.train.Saver(tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, "d_gm0_fmnist_10"))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    train_features(sess=sess, feature_model=f, samplers=s, num_steps=40000, keep_prob=0.8, f_params=f_params)
    fs.save(sess, "./models/fmnist/d_gm0_fmnist_10.ckpt")
```

```python
tf.reset_default_graph()
f, f_params = get_feature_model(feature_model=GCNNMedium1, settings=Triplet_Distance_Settings,
                                num_features=10, data_format=[28, 28, 1],
                                f_lr=0.02, f_reg=0.001, f_scope="d_gm1_fmnist_10")
fs = tf.train.Saver(tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, "d_gm1_fmnist_10"))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    train_features(sess=sess, feature_model=f, samplers=s, num_steps=40000, keep_prob=0.8, f_params=f_params)
    fs.save(sess, "./models/fmnist/d_gm1_fmnist_10.ckpt")
```
