```python
%load_ext autoreload
%autoreload 2
import numpy as np
import tensorflow as tf
from features.settings import Baseline_Settings, Triplet_Hinge_Settings, Triplet_Log_Settings, Pair_Hinge_Settings
from features.models import GCNNSmall0, GCNNMedium0, GCNNMedium1
from learn import get_feature_model, train_features, test_features
from datasets import load_fashion_mnist
```

# 10

```python
tf.reset_default_graph()
f, f_params = get_feature_model(feature_model=GCNNSmall0, settings=Baseline_Settings,
                                num_features=10, data_format=[28, 28, 1],
                                f_lr=0.02, f_reg=0.001, f_scope="b_gs0_fmnist_10")
fs = tf.train.Saver(tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, "b_gs0_fmnist_10"))

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    train_features(sess=sess, feature_model=f, samplers=s, num_steps=40000, keep_prob=0.8, f_params=f_params)
    fs.save(sess, "./models/fmnist/b_gs0_fmnist_10.ckpt")
```
